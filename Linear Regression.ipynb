```python
# Q5 How is it varying with 'sepal length' and 'sepal width'?
y=iris[['sepal_length']]
```

```python
x=iris[['sepal_width']]
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```
